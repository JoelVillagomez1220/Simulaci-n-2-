Procedimiento a seguir para realizar la prueba de bondad de ajuste

- Planteamiento de Hiptesis:

Define tus hiptesis nula (H0) y alternativa (H1):
- **H0:** La muestra sigue una distribucin especfica.
- **H1:** La muestra no sigue la distribucin especfica.

- Eleccin de la Distribucin:

Elige la distribucin que crees que sigue tu conjunto de datos.

- Recopilacin de Datos:

Obtn una muestra aleatoria de datos y organzalos en clases o categoras.

- Establecimiento de Intervalos de Clases:

Divide el rango de tus datos en intervalos de clases. Asegrate de que cada dato caiga en una clase especfica.

- Frecuencias Observadas y Esperadas:

Registra las frecuencias observadas para cada clase. Estas son las frecuencias reales de tu muestra.

Calcula las frecuencias esperadas para cada clase utilizando la distribucin terica seleccionada. Puedes hacer esto mediante clculos o tablas de probabilidad.

- Clculo de la Estadstica Chi-Cuadrada:

\[ \chi^2 = \sum \frac{(O_i - E_i)^2}{E_i} \]

Donde:
- \( \chi^2 \) es la estadstica chi-cuadrada.
- \( O_i \) es la frecuencia observada para la clase \( i \).
- \( E_i \) es la frecuencia esperada para la clase \( i \).

- Grados de Libertad

- Consulta la Tabla Chi-Cuadrada:

Utiliza la estadstica chi-cuadrada calculada y los grados de libertad para consultar la tabla chi-cuadrada. Determina el valor crtico que te permitir comparar con tu estadstica calculada.
Si el valor calculado es mayor que el valor crtico, rechaza la hiptesis nula.


```python
#Ejercicio 3

#Hiptesis nula (H0): Los datos siguen una distribucin uniforme.
#Hiptesis alternativa (H1): Los datos no siguen una distribucin uniforme.

import numpy as np
from scipy.stats import chi2

# Datos proporcionados
data = np.array([38, 33, 25, 5, 69, 35, 98, 52, 12, 79,
                 50, 46, 95, 42, 49, 11, 78, 34, 2, 43,
                 7, 50, 5, 91, 77, 18, 21, 4, 17, 62,
                 91, 36, 48, 88, 52, 76, 99, 73, 82, 90])

# Definir intervalos
num_intervals = 10
interval_limits = np.linspace(min(data), max(data), num_intervals+1)

# Contar frecuencia observada en cada intervalo
observed_freq, _ = np.histogram(data, bins=interval_limits)

# Calcular frecuencia esperada
expected_freq = len(data) / num_intervals

# Calcular chi^2
chi_squared = np.sum((observed_freq - expected_freq)**2 / expected_freq)

# Grados de libertad
df = num_intervals - 1

# Valor crtico para un nivel de significancia del 95%
alpha = 0.05
critical_value = chi2.ppf(1 - alpha, df)

# Comparar con valor crtico
if chi_squared > critical_value:
    print(f'Rechazar la hiptesis nula. Chi^2 = {chi_squared} > Valor Crtico = {critical_value}')
else:
    print(f'No se puede rechazar la hiptesis nula. Chi^2 = {chi_squared} <= Valor Crtico = {critical_value}')
```
