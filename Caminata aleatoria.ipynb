```python
#Villagómez de Alba Joel Ricardo
import turtle
import random



# Función para realizar una caminata aleatoria con un número dado de pasos y devolver las coordenadas finales
def caminata(steps):
    x, y = 0, 0  # Coordenadas iniciales
    for _ in range(steps):
        direction = random.choice([ "Este", "Oeste"])
        if direction == "Norte":
            y += 1
        elif direction == "Sur":
            y -= 1
        elif direction == "Este":
            x += 1
        elif direction == "Oeste":
            x -= 1
    return x, y

# Función para calcular la distancia Manhattan entre dos puntos
def distancia(x1, y1, x2, y2):
    return abs(x2 - x1) + abs(y2 - y1)

numero_caminatas = 1000

for longitud in range(1, 100):
    exitos = 0
    for _ in range(numero_caminatas):
        x, y = caminata(longitud)
        distancia_total = distancia(0, 0, x, y)
        if distancia_total <= 4:
            exitos += 1
    porcentaje = 100 * (exitos / numero_caminatas)
    print("Pasos =", longitud, ", % exitos =", porcentaje)
```
