Procedimiento a seguir para realizar la prueba de bondad de ajuste

- Planteamiento de Hipótesis:

Define tus hipótesis nula (H0) y alternativa (H1):
- **H0:** La muestra sigue una distribución específica.
- **H1:** La muestra no sigue la distribución específica.

- Elección de la Distribución:

Elige la distribución que crees que sigue tu conjunto de datos.

- Recopilación de Datos:

Obtén una muestra aleatoria de datos y organízalos en clases o categorías.

- Establecimiento de Intervalos de Clases:

Divide el rango de tus datos en intervalos de clases. Asegúrate de que cada dato caiga en una clase específica.

- Frecuencias Observadas y Esperadas:

Registra las frecuencias observadas para cada clase. Estas son las frecuencias reales de tu muestra.

Calcula las frecuencias esperadas para cada clase utilizando la distribución teórica seleccionada. Puedes hacer esto mediante cálculos o tablas de probabilidad.

- Cálculo de la Estadística Chi-Cuadrada:

\[ \chi^2 = \sum \frac{(O_i - E_i)^2}{E_i} \]

Donde:
- \( \chi^2 \) es la estadística chi-cuadrada.
- \( O_i \) es la frecuencia observada para la clase \( i \).
- \( E_i \) es la frecuencia esperada para la clase \( i \).

- Grados de Libertad

- Consulta la Tabla Chi-Cuadrada:

Utiliza la estadística chi-cuadrada calculada y los grados de libertad para consultar la tabla chi-cuadrada. Determina el valor crítico que te permitirá comparar con tu estadística calculada.
Si el valor calculado es mayor que el valor crítico, rechaza la hipótesis nula.


```python
#Ejercicio 3

#Hipótesis nula (H0): Los datos siguen una distribución uniforme.
#Hipótesis alternativa (H1): Los datos no siguen una distribución uniforme.

import numpy as np
from scipy.stats import chi2

# Datos proporcionados
data = np.array([38, 33, 25, 5, 69, 35, 98, 52, 12, 79,
                 50, 46, 95, 42, 49, 11, 78, 34, 2, 43,
                 7, 50, 5, 91, 77, 18, 21, 4, 17, 62,
                 91, 36, 48, 88, 52, 76, 99, 73, 82, 90])

# Definir intervalos
num_intervals = 10
interval_limits = np.linspace(min(data), max(data), num_intervals+1)

# Contar frecuencia observada en cada intervalo
observed_freq, _ = np.histogram(data, bins=interval_limits)

# Calcular frecuencia esperada
expected_freq = len(data) / num_intervals

# Calcular chi^2
chi_squared = np.sum((observed_freq - expected_freq)**2 / expected_freq)

# Grados de libertad
df = num_intervals - 1

# Valor crítico para un nivel de significancia del 95%
alpha = 0.05
critical_value = chi2.ppf(1 - alpha, df)

# Comparar con valor crítico
if chi_squared > critical_value:
    print(f'Rechazar la hipótesis nula. Chi^2 = {chi_squared} > Valor Crítico = {critical_value}')
else:
    print(f'No se puede rechazar la hipótesis nula. Chi^2 = {chi_squared} <= Valor Crítico = {critical_value}')
```
